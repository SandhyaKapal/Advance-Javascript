```python
// Create a new Map
let myMap = new Map();

// 1️⃣ Add key-value pairs to the Map
myMap.set("name", "Sandhya");
myMap.set("age", 22);
myMap.set("city", "Mumbai");

// 2️⃣ Check if a specific key exists
console.log(myMap.has("age"));       // true
console.log(myMap.has("country"));   // false

// 3️⃣ Retrieve the value associated with a given key
console.log(myMap.get("name"));      // Sandhya
console.log(myMap.get("city"));      // Mumbai

// 4️⃣ Iterate through all key-value pairs
for (let [key, value] of myMap) {
    console.log(key + " : " + value);
}
```

Problem 2

Create a Map to store contact information (name, age, email, location) and implement a function to retrieve
contact details by name.

```python
// Create a Map to store contact information
let contacts = new Map();

// Add contact details
contacts.set("Sandhya", {
    age: 22,
    email: "sandhya@example.com",
    location: "Mumbai"
});

contacts.set("Rahul", {
    age: 25,
    email: "rahul@example.com",
    location: "Pune"
});

contacts.set("Neha", {
    age: 28,
    email: "neha@example.com",
    location: "Delhi"
});

// Function to retrieve contact details by name
function getContactDetails(name) {
    if (contacts.has(name)) {
        return contacts.get(name);
    } else {
        return "Contact not found!";
    }
}

// Test the function
console.log(getContactDetails("Rahul"));
console.log(getContactDetails("Sandhya"));
console.log(getContactDetails("Amit")); // Not
```

You are given an array of numbers named myArray. Create a function that takes any number of arguments
and adds them to the existing array. Use the spread and rest operator.

```python
// Given array
let myArray = [10, 20, 30];

// Function to add any number of arguments to the array
function addToArray(...nums) {
    // Using spread operator to merge
    myArray = [...myArray, ...nums];
}

// Test the function
addToArray(40, 50);
addToArray(60, 70, 80);

console.log(myArray);
```
